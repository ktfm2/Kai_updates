```python
import agama
import matplotlib.pyplot as plt
import numpy as np
```

```python
pot = agama.Potential(type = 'Spheroid',gamma=1,beta=3.1, outerCutoffRadius=50)
```

```python
#agama.Potential?
```

```python
df = agama.DistributionFunction(type='QuasiSpherical',potential=pot)
```

```python
model=agama.GalaxyModel(pot,df)
```

```python
M= model.sample(100000)
```

```python
plt.plot(M[0][:,0],M[0][:,1],'.',ms=1) 
#plt.xlim(-10,10)
#plt.ylim(-10,10)
plt.gca().set_aspect('equal')
```

```python
plt.hist(M[0][:,0],range=[-10,10])
```

```python
plt.hist(M[0][:,1],range=[-10,10])
```
